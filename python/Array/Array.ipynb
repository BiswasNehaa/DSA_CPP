# Second Largest

```python
def sec_largest(arr):
    large=seclarge=arr[0]
    
    for i in range(len(arr)):
        if arr[i]>large :
            seclarge=large
            large=arr[i]
        elif arr[i]>seclarge and arr[i]!=large:
            seclarge=arr[i]
        
    print(seclarge)

sec_largest([2,6,1,5])
sec_largest([5, 5, 3])
```
